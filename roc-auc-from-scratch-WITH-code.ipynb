# ROC

- **R**eceiver **O**perating **C**haracteristic.

- [Wiki:](https://en.wikipedia.org/wiki/Receiver_operating_characteristic) It's a graphical plot used to compare the diagnostic ability of a binary classifier. 

- The name comes from its use in military radar receivers.

- Plot of *true positive rate* (TPR) vs *false positive rate* (FPR) at various thresholds. 

- Example: 
<p><a href="https://commons.wikimedia.org/wiki/File:Roccurves.png#/media/File:Roccurves.png"><img src="https://upload.wikimedia.org/wikipedia/commons/6/6b/Roccurves.png" alt="Roccurves.png"></a><br>By <a href="https://en.wikipedia.org/wiki/User:BOR" class="extiw" title="en:User:BOR">BOR</a> at the <a href="https://en.wikipedia.org/wiki/" class="extiw" title="w:">English-language Wikipedia</a>, <a href="http://creativecommons.org/licenses/by-sa/3.0/" title="Creative Commons Attribution-Share Alike 3.0">CC BY-SA 3.0</a>, <a href="https://commons.wikimedia.org/w/index.php?curid=10714489">Link</a></p>

- Alternative names for the axis components:

| x-axis | y-axis |
|----|----|
| True positive rate (TPR) | False Positive Rate (FPR) |
| Sensitivity | $1 -$ specificity |
| Recall | $1 -$ precision |
| hit rate | fall out |

- Its usefulness comes from the fact that this can be run independantly of the *cost calculation*/ the *class distribution*.
- Related to cost/benifit analysis of diagnostic decision making.
