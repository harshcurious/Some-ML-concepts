# What are ERS?
Dynamic power transmission via overhead lines (trucks, high power), conductive rails embedded in the asphalt (cars and trucks, medium power), or induction coils  beneath the asphalt (cars and trucks, lower power, no mechanical connection).

The following business case is based solely on truck traffic.

# How can society benefit from ERS technology?

*   Eliminating range anxiety without relying on lithium by reducing the required battery capacities
*   Consuming electricity the moment it is generated by solar and wind power, rather than in the morning and evening.
*   Less need for energy storage and certainly no need for additional nuclear power plants
*   Lower overall energy demand and greater grid stability compared to static charging
*   Faster rollout compared to static charging due to fewer required grid connections
*   More cost-effective overall than static charging, thereby also reducing the burden on taxpayers (The ICCT estimates that the cost of electrification in Germany for heavy-duty transport alone would amount to !46 billion! euros)

# How can even Deutsche Bahn benefit from ERS?

*   Local passenger rail transport in Germany is only 28% utilized on average. Even a car achieves 33%.
*   At the same time, however, local passenger rail transport accounts for 61% of network capacity (a huge waste of expensive infrastructure)
*   Long-distance passenger rail transport accounts only for 13% network capacity and is also slowed down by local transport.
*   If local rail passenger transport were to be eliminated, the capacity of long-distance transport could be increased by a factor of at least 5.7 (13% + 61%) / 13%. In practice, the achievable capacity would be significantly higher due to the more homogeneous speeds on the network and the elimination of
overtaking maneuvers without corresponding passing opportunities.
*   Japan and France have separate high-speed networks, which is why their rail system works well.The planned “Integraler Taktfahrplan” in Germany (“Deutschlandtakt”), whose implementation had to be postponed for several decades, will not attract significantly more passengers because it hardly reduces the time required for long distances and continues to rely on completely inefficient local transport. In Switzerland, which serves as a model for this, the “Integraler Taktfahrplan” only works because of the short distances of a small country and the relatively low speed of long-distance transport that this allows.

Therefore, the obvious **solution** would be to **partially replace local and freight transport with ERS** to make room for the Intercity Express (ICE). This could **allow the ICE to become a mobility solution**—as in Japan—that **domestic flights cannot compete with.**

```python
# ERS -- Optimize
import seekerdemo as skr

s = {
    "charging_factor": {"center":_fv(wcc),"dist":wcd.value,"p1":_fv(wcp1),"p2":_fv(wcp2)},
    "sell_price":      {"center":_fv(wsc),"dist":wsd.value,"p1":_fv(wsp1),"p2":_fv(wsp2)},
    "elec_price":      {"center":_fv(wec),"dist":wed.value,"p1":_fv(wep1),"p2":_fv(wep2)},
    "capex":           {"center":_fv(wxc),"dist":wxd.value,"p1":_fv(wxp1),"p2":_fv(wxp2)},
    "wacc":            {"center":_fv(wwc),"dist":wwd.value,"p1":_fv(wwp1),"p2":_fv(wwp2)},
    "diff_base":       {"center":_fv(wdbc),"dist":wdbd.value,"p1":_fv(wdbp1),"p2":_fv(wdbp2)},
    "diff_max":        {"center":_fv(wdmc),"dist":wdmd.value,"p1":_fv(wdmp1),"p2":_fv(wdmp2)},
    "diff_turn":       {"center":float(wdtc_int.value),"dist":wdtd.value,"p1":_fv(wdtp1),"p2":_fv(wdtp2)},
    "diff_growth":     {"center":_fv(wdgc),"dist":wdgd.value,"p1":_fv(wdgp1),"p2":_fv(wdgp2)},
}

# Effective central values for reporting
def _eff_ctr(c,d,p1,p2):
    if d=='Uniform':
        if p1 and not p2:   return (min(c,p1)+max(c,p1))/2
        elif not p1 and p2: return (min(c,p2)+max(c,p2))/2
        elif p1 and p2:     return (min(p1,p2)+max(p1,p2))/2
    return c

skw_live={}
for cn,sv in skw.items():
    try: skw_live[cn]=float(sv.value)
    except: skw_live[cn]=0.9

cfg_countries=[]
for cn,cw in cww.items():
    if not cw["cb"].value: continue
    dcn=mp_name_w[cn].value.strip() if cn in mp_name_w and mp_name_w[cn].value.strip() else cn
    real_hw=_fv(hw_w.get(cn,txt("0")))
    if real_hw<=0: print(f"no highway data for {cn} -- skipped"); continue
    mkm=max(100, int(round(real_hw*0.5/100)*100))
    sig=skw_live.get(cn,0.9)
    try: dom=float(str(mp_dom_w[cn].value).replace(",",""))
    except: dom=0.0
    try: cb2=float(str(mp_cb_w[cn].value).replace(",",""))
    except: cb2=0.0
    cb_p_center=_eff_ctr(_fv(cw["ctr"]),cw["dist"].value,_fv(cw["p1"]),_fv(cw["p2"]))
    cb_p_p1=_fv(cw["p1"]); cb_p_p2=_fv(cw["p2"])
    cb_p_dist=cw["dist"].value
    cb_p_dict={"dist":cb_p_dist,"center":cb_p_center,"p1":cb_p_p1,"p2":cb_p_p2}
    cb_prob=cb_p_center
    total_mp=dom+cb_prob*cb2
    vol_list=[0.0]
    for km in range(100,int(mkm)+1,100):
        anteil=min(max(km/mkm,1e-7),1-1e-7)
        pen=float(np.clip(1-_norm.cdf(_norm.ppf(1-anteil)-sig**2),0,1))
        vol_list.append(pen*total_mp)
    while len(vol_list)<62: vol_list.append(vol_list[-1])
    cfg_countries.append({"name":dcn,"orig_name":cn,"max_km":mkm,"sigma":sig,
                           "vol_list":vol_list,"total_mp":total_mp})

results=[]; years=list(range(wms.value,wme.value+1)); NY=len(years)
stop_idx=wsy.value-wms.value

capex_ctr = _eff_ctr(_fv(wxc),wxd.value,_fv(wxp1),_fv(wxp2))
wacc_ctr  = _eff_ctr(_fv(wwc),wwd.value,_fv(wwp1),_fv(wwp2))
sell_ctr  = _eff_ctr(_fv(wsc),wsd.value,_fv(wsp1),_fv(wsp2))
elec_ctr  = _eff_ctr(_fv(wec),wed.value,_fv(wep1),_fv(wep2))
ch_ctr    = _eff_ctr(_fv(wcc),wcd.value,_fv(wcp1),_fv(wcp2))
maint_pct = _fv(wmp_)
fin_years=way.value; sys_eff_v=_fv(wse); truck_kwh_v=_fv(wtk)

def _mk(env, p):
    d2,c,p1,p2=p["dist"],p["center"],p["p1"],p["p2"]
    if d2=="Uniform":
        if not p1 and not p2:
            # No bounds: deterministic
            return env.continuous_uniform(c*0.9999, c*1.0001)
        elif p1 and not p2:
            # Only lower bound given: uniform between lower and center
            lo,hi=min(c,p1),max(c,p1)
            if lo>=hi: return env.continuous_uniform(c*0.9999, c*1.0001)
            return env.continuous_uniform(lo,hi)
        elif not p1 and p2:
            # Only upper bound given: uniform between center and upper
            lo,hi=min(c,p2),max(c,p2)
            if lo>=hi: return env.continuous_uniform(c*0.9999, c*1.0001)
            return env.continuous_uniform(lo,hi)
        else:
            # Both bounds given: uniform between them (ignore center)
            lo,hi=min(p1,p2),max(p1,p2)
            if lo>=hi: return env.continuous_uniform(c*0.9999, c*1.0001)
            return env.continuous_uniform(lo,hi)
    elif d2=="Normal":
        if not p1:
            return env.continuous_uniform(c*0.9999, c*1.0001)
        sd=p1
        return env.normal(c,sd,max(0,c-4*sd),c+4*sd)
    elif d2=="Gamma (right-skewed)":
        if not p1 or not p2:
            return env.continuous_uniform(c*0.9999, c*1.0001)
        sh,sc=p1,p2
        return env.gamma(sh*sc,sh*sc*3,sh*sc)
    else:
        return env.continuous_uniform(c*0.9999, c*1.0001)

for country in cfg_countries:
    cn=country["name"]; vol_list=country["vol_list"]; mkm=country["max_km"]
    print(f"Optimising {cn}...")
    env=skr.Env("",True)

    ch  = _mk(env, s["charging_factor"])
    sp2 = _mk(env, s["sell_price"])
    ep  = _mk(env, s["elec_price"])
    cf  = _mk(env, s["capex"])
    wc2 = _mk(env, s["wacc"])
    db  = _mk(env, s["diff_base"])
    dm  = _mk(env, s["diff_max"])
    dt  = _mk(env, s["diff_turn"])
    dg  = _mk(env, s["diff_growth"])
    cb_p_stoch=_mk(env, cb_p_dict)
    # total_mp_stoch: stochastic market potential based on CB prob distribution
    total_mp_stoch=env.convert(dom)+cb_p_stoch*env.convert(cb2)

    try: _max_b=int(str(wmax_build.value).replace(",",""))
    except: _max_b=500
    b=[env.ordinal(0,min(_max_b,int(mkm))) for _ in range(stop_idx+1)]
    b+=[env.convert(0)]*(NY-stop_idx-1)
    total=[b[0]]
    for i in range(1,NY): total.append(env.min([total[-1]+b[i],env.convert(mkm)]))

    # vol_list built with central CB prob; scale entries to unit penetration
    # then multiply by stochastic total_mp
    total_mp_center=dom+cb_p_center*cb2
    if total_mp_center>0:
        pen_list=[v/total_mp_center for v in vol_list]
    else:
        pen_list=[0.0]*len(vol_list)
    pen_ten=env.tensor([pen_list]*NY)
    ind2=[env.div(t,100) for t in total]
    mod2=[(t%100)/100 for t in total]
    diff=[db+(dm-db)/(1+env.exp(-dg*(env.convert(y)-dt))) for y in range(NY)]
    net_rpt=(sp2-ep)/sys_eff_v*truck_kwh_v*ch

    revenue=[
        ((1-mod2[y])*pen_ten[[env.convert(y),ind2[y]]]+
          mod2[y]*pen_ten[[env.convert(y),env.min([ind2[y]+1,env.convert(len(pen_list)-1)])]])
        *total_mp_stoch*diff[y]*net_rpt
        for y in range(NY)
    ]


    # Exact stochastic annuity via lookup table
    _wlo=min(s["wacc"]["p1"] if s["wacc"]["p1"] else s["wacc"]["center"],
             s["wacc"]["p2"] if s["wacc"]["p2"] else s["wacc"]["center"])
    _whi=max(s["wacc"]["p1"] if s["wacc"]["p1"] else s["wacc"]["center"],
             s["wacc"]["p2"] if s["wacc"]["p2"] else s["wacc"]["center"])
    if _wlo==_whi: _wlo=_wlo*0.9999; _whi=_whi*1.0001
    _N=20
    _ann_list=[(_wlo+(_whi-_wlo)*i/_N)/(1-(1+(_wlo+(_whi-_wlo)*i/_N))**(-fin_years))
               for i in range(_N+1)]
    _widx=env.ordinal(0,_N)
    ann_rate_stoch=env.tensor([_ann_list])[[env.convert(0),_widx]]
    yearly_costs=[]
    for y in range(NY):
        mc=total[y]*cf*maint_pct
        ac=env.convert(0)
        for t in range(min(y+1,stop_idx+1)):
            if y-t<fin_years: ac=ac+b[t]*cf*ann_rate_stoch
        yearly_costs.append(mc+ac)

    total_costs=env.sum(yearly_costs)
    profit=env.sum(revenue)-total_costs
    ep2=env.aggregate_mean(profit)
    cvar=env.aggregate_cvar(profit,0.05,False)   # CVaR: expected value of worst 5%
    env.set_report(2,[ep2/1e6,cvar/1e6],["E_profit_M","CVaR5_M"])
    _wv=w_weight.value if w_obj.value=="weighted" else 0.5
    if w_obj.value=="mean":
        _obj=ep2/1e6
    elif w_obj.value=="weighted":
        _obj=(_wv*ep2+(1-_wv)*cvar)/1e6
    else:
        _obj=(ep2+cvar)/1e6
    env.maximize(_obj,10)
    env.evaluate()

    ob=[int(b[t].get_value()) for t in range(NY)]
    oc=list(np.cumsum(ob))
    mu=ep2.get_value()    # E[total profit] in EUR
    q5=cvar.get_value()    # CVaR 5%: expected value of worst 5% scenarios

    # Deterministic annual breakdown using central parameter values
    rv=[]; cv=[]; dv=[]
    for t in range(NY):
        km_built=oc[t]
        idx=min(int(km_built//100),len(vol_list)-1)
        frac=(km_built%100)/100
        vol=vol_list[idx]*(1-frac)+vol_list[min(idx+1,len(vol_list)-1)]*frac
        b_val=_fv(wdbc); m_val=_fv(wdmc); turn_val=wdtc_int.value; g_val=_fv(wdgc)
        diff_val=b_val+(m_val-b_val)/(1+math.exp(-g_val*(t-turn_val)))
        net_r=(sell_ctr-elec_ctr)/_fv(wse)*_fv(wtk)*ch_ctr
        rv.append(vol*diff_val*net_r)
        ann_r=wacc_ctr/(1-(1+wacc_ctr)**(-fin_years))
        mc=km_built*capex_ctr*maint_pct
        ac=sum(capex_ctr*ann_r*ob[s2] for s2 in range(min(t+1,stop_idx+1)) if t-s2<fin_years)
        cv.append(mc+ac)
        dv.append(diff_val)
    op_raw=[rv[t]-cv[t] for t in range(NY)]
    # Scale annual values so cumulative total matches Seeker E[profit]
    op_sum=sum(op_raw)
    scale_f=mu/op_sum if abs(op_sum)>1e6 else 1.0
    op=[v*scale_f for v in op_raw]
    rv=[v*scale_f for v in rv]
    cv=[rv[t]-op[t] for t in range(NY)]

    # Std from Q05: Q05 = mu - 1.645*sigma  =>  sigma = (mu - Q05) / 1.645
    # CVaR <= mean always; sg from CVaR distance
    sg=abs(mu-q5)/2.063 if mu>q5 else abs(mu)*0.1  # 2.063 = E[Z|Z<-1.645] factor

    results.append({"name":cn,"exp_profit_total":mu,"q05_profit":q5,
        "opt_infra_km":oc[-1],"mean_profit":mu,"std_profit":sg,
        "min_profit":min(op),"max_profit":max(op),"skewness":0.3,
        "annual":{"new_build":ob,"cumulative":oc,"exp_revenue":rv,
                  "exp_costs":cv,"exp_profit":op,"diffusion":dv}})
    env.end()
    print(f"  E[profit]={mu/1e6:.0f}M EUR  CVaR={q5/1e6:.0f}M EUR  opt_infra={oc[-1]:.0f}km")

chart_images={}
for res in results:
    cn=res["name"]; ann=res["annual"]
    fig=plt.figure(figsize=(16,7))
    fig.suptitle(cn,fontsize=13,fontweight="bold",color="#1F4E79")
    g=mgs.GridSpec(2,3,figure=fig,hspace=.42,wspace=.32)

    ax=fig.add_subplot(g[0,0])
    rv_b=[v/1e9 for v in ann["exp_revenue"]]
    cv_b=[v/1e9 for v in ann["exp_costs"]]
    op_b=[v/1e9 for v in ann["exp_profit"]]
    ax.plot(years,rv_b,color="#2E75B6",lw=2,label="Revenue")
    ax.plot(years,cv_b,color="#E74C3C",lw=2,label="Costs")
    ax.plot(years,op_b,color="#8E44AD",lw=2,label="Net profit")
    ax.axhline(0,color="k",lw=.5); ax.set_title("Annual financials (B EUR)"); ax.legend(fontsize=8)

    ax=fig.add_subplot(g[0,1])
    cp=[sum(op_b[:i+1]) for i in range(len(op_b))]
    ax.plot(years,cp,color="#8E44AD",lw=2)
    ax.fill_between(years,cp,alpha=0.1,color="#8E44AD")
    ax.axhline(0,color="k",lw=.5,ls="--"); ax.set_title("Cumulative net profit (B EUR)")

    ax=fig.add_subplot(g[0,2])
    ax.bar(years,ann["new_build"],color="#2E75B6",alpha=.7,label="New km")
    ax.plot(years,ann["cumulative"],color="#1F4E79",lw=2,label="Cumulative km")
    ax.set_title("Infrastructure build-out (km)"); ax.legend(fontsize=8)

    ax=fig.add_subplot(g[1,0])
    mu_b=res["mean_profit"]/1e9; sg_b=res["std_profit"]/1e9
    if sg_b>0:
        q5_b=res["q05_profit"]/1e9
        left=min(mu_b-4*sg_b, q5_b-abs(sg_b)*0.5)
        x_rng=np.linspace(left, mu_b+2*sg_b, 300)
        y_dens=_norm.pdf(x_rng,mu_b,sg_b)
        ax.plot(x_rng,y_dens,color="#8E44AD",lw=2)
        ax.fill_between(x_rng,y_dens,alpha=.12,color="#8E44AD")
        ax.axvline(q5_b,color="#E74C3C",lw=1.5,ls="--",label=f"CVaR 5%: {q5_b:.2f}B")
        ax.axvline(mu_b,color="#27AE60",lw=1.5,ls="--",label=f"Mean: {mu_b:.2f}B")
        ax.legend(fontsize=8)
    ax.set_xlabel("Total profit (Billion EUR)"); ax.set_title("Profit distribution (B EUR)")
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda v,_: f"{v:.3f}"))

    ax=fig.add_subplot(g[1,1]); ax.axis("off")
    for yi,(lbl2,val) in enumerate([
        ("E[profit]",  f"{res['exp_profit_total']/1e9:.2f} B EUR"),
        ("CVaR 5%",    f"{res['q05_profit']/1e9:.2f} B EUR"),
        ("Opt. infra", f"{res['opt_infra_km']:,.0f} km"),
        ("Std dev",    f"{res['std_profit']/1e9:.2f} B EUR"),
    ]):
        y2=.88-yi*.22
        ax.text(.05,y2,lbl2,transform=ax.transAxes,fontsize=9,color="#595959")
        ax.text(.97,y2,val,transform=ax.transAxes,fontsize=10,fontweight="bold",color="#1F4E79",ha="right")
    ax.set_title("Key metrics")

    ax=fig.add_subplot(g[1,2]); ax.axis("off")
    plt.tight_layout()
    buf2=io.BytesIO(); fig.savefig(buf2,format="png",dpi=120,bbox_inches="tight")
    buf2.seek(0); chart_images[cn]=buf2
    plt.show()

print(f"\n{len(results)} countries optimised")
```
